```python
import matplotlib.pyplot as plt

train_gen = data["train_gen"]
if train_gen is not None:
    images, labels = next(train_gen)
    plt.figure(figsize=(12, 8))
    for i in range(9):
        plt.subplot(3, 3, i + 1)
        plt.imshow(images[i])
        plt.title(f"Label: {labels[i]}")
        plt.axis("off")
    plt.show()
else:
    print("No image generator found.")
```
